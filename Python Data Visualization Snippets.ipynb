```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot(range(7), linestyle='-', marker='o', color='b', label='x')
squares = [x**2 for x in range(7)]
plt.plot(squares, linestyle='--', marker='o', color='r', label='x^2') 
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
